## Beispieldaten

```python
def beispieldaten(anz=10, nan_age=True):
    
    # Beispieldaten
    np.random.seed(11)
    df = pak.people(anz)
    df = pak.drop_cols(df,['features','history'])
    
    # translatedaten
    # mit zwei NaNs
    tr = df.tail(int(anz/2))
    tr = pak.drop_cols(tr, ['birthplace'])
    tr.secret = '##' + tr.secret + '##'
    tr.postal_code      = -tr.postal_code
    if nan_age:
        tr.iloc[1,1] = np.NaN
    tr.iloc[3,4] = None
    tr = pak.reset_index(tr)
    return  df, tr

df,tr = beispieldaten()

df = pak.change_datatype(df, verbose=False)
df.age = pak.change_datatype(df.age, 'float32')
df = pak.rename_col(df, 'age','aGe')
df
```

## pak.update_col
* Die Datensätze 5, 6, 7, 8, 9 sind in beiden DataFrames vorhanden. Aber:
* NaN oder None verhindern die Zuordnung, daher wird Datensatz 6 (Sarah) nicht geschrieben
* NaN oder None werden nicht übertragen,  daher wird Datensatz 8 (Pia)   nicht geschrieben
* Es bleiben also 5, 7, 9 (Sigrid, Arthur, Elena)

```python
df1 = df.copy()
df1 = pak.update_col( df1, tr, left_on=['first_name','aGe'], right_on=['first_name','age'], col='secret', keep='_alt')
pak.get_different_rows(df, df1)
```

## Mit mask

```python
df2 = df1.copy()
agesschwelle = int(df2.aGe.mean()+1)
print('agesschwelle',agesschwelle)
mask = (df2.aGe < agesschwelle)
df2[mask] = pak.update_col( df2[mask], tr, left_on=['first_name','aGe'], right_on=['first_name','age'], col='postal_code')
```
