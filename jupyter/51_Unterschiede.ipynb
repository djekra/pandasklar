## isin
über mehrere Spalten

```python
?pak.isin
```

```python
df1 = pak.dataframe( [ list('Würmchen'), 
                       list('Wurmloch'), 
                       list('Witzbold'), 
                       list('Heuernte'), 
                       list('Halbgott'), 
                       list('Halbzeit'), 
                       list('Hausfrau'),       
                       list('Hebelarm'),                     
                ] )

df2 = pak.dataframe( [ list('Babykorb'), 
                       list('abfällig'), 
                       list('Abgründe'), 
                       list('Kätzchen'), 
                       list('Landwirt'), 
                       list('Handkuss'),  # hier gibt es eine Übereinstimmung
                       list('Saugrohr'),       
                       list('Trugbild'),                     
                ] )

df2
```

```python
mask = pak.isin(df1, df2, on=['A','B'])
df1[mask]
```

```python
mask = ~pak.isin(df1, df2, on=['A','B'])
df1[mask]
```

## check_identical
Vergleich zweier DataFrames

```python
?pak.check_identical
```

```python
df1 = pak.dataframe( [ list('Babykorb'), 
                       list('abfällig'), 
                       list('Abgründe'), 
                       list('Kätzchen'), 
                       list('Landwirt'), 
                       list('lebendig'), 
                       list('Saugrohr'),       
                       list('Trugbild'),                     
                ] )

df2 = pak.dataframe( [ list('Babykorb'), 
                       list('abfällig'), 
                       list('Abgründe'), 
                       list('Kätzchen'), 
                       list('Landwirt'), 
                       list('lebendig'), 
                       list('Saugrohr'),       
                       list('Trugbild'),                     
                ] )

df1
```

```python
# Anfangs sind die DataFrames identisch
assert pak.check_identical(df1, df2, verbose=True)
```

```python
# Eine Änderung >> nicht identisch
mask = df2['A'] == 'L'
df2.loc[mask,'A'] = 'R'
assert not pak.check_identical(df1, df2, verbose=True)
```

```python
# Zurückändern >> wieder identisch
mask = df2['A'] == 'R'
df2.loc[mask,'A'] = 'L'
assert pak.check_identical(df1, df2, verbose=True)
```

```python
# Spaltenreihenfolge und Zeilenreihenfolge werden geändert
df2 = pak.move_cols(df2,'D').sort_values('D')
df2
```

```python
df2
```

```python
# immer noch identisch
assert pak.check_identical(df1, df2, verbose=True)
```

## get_different_rows
Vergleich zweier DataFrames

```python
?pak.get_different_rows
```

```python
df1 = pak.dataframe( [ list('Babykorb'), 
                       list('abfällig'), 
                       list('Abgründe'), 
                       list('Kätzchen'), 
                       list('Landwirt'), 
                       list('lebendig'), 
                       list('Saugrohr'),       
                       list('Trugbild'),                     
                ] )

df2 = pak.dataframe( [ list('Babykorb'), 
                       list('abfällig'), 
                       list('Abgründe'), 
                       list('Kätzchen'), 
                       list('Landwirt'), 
                       list('lebendig'), 
                       list('Saugrohr'),       
                       list('Trugbild'),                     
                ] )

df1
```

```python
# Anfangs sind beide DataFrames identisch
pak.get_different_rows(df1, df2)
```

```python
# Spalten und Zeilen umsortieren
df2 = pak.move_cols(df2,'D').sort_values('D')
df2
```

```python
# >> immer noch keine Zeilen, die sich unterscheiden
pak.get_different_rows(df1, df2)
```

```python
# Einstreufehler
df1['B'] = df1.B.apply( pak.decorate, p=0.2, special='#' )
df1
```

```python
# Zeilen anzeigen, die sich jetzt unterscheiden
pak.get_different_rows(df1, df2)
```
