```python
dupmask = df1.duplicated(subset=['vorname','alter'], keep=False) # keep=False bedeutet keep='both'. keep kann auch 'first' oder 'last' sein.
#dupmask
```

```python
# die Duplikate
df1[dupmask].sort_values(['vorname','alter'])
```
