# startswith

```python
?pak.fast_startswith
```

```python
# create data to search in
size = 10000
manystrings = pd.DataFrame(pak.random_series(size*10,'string',len_min=10, len_max=10, name='U'))

# create data to search
search_me   = pd.DataFrame(pak.random_series(size,'string',len_min=2, len_max=4, name='V'))

#analyse_cols(search_me)
```

```python
%%time 
mask = manystrings.U.str.startswith(tuple(search_me.V))
result0 = manystrings[mask]
# result0: built-in startswith
# Wall time: 1min 25s (Notepad, size = 10000)
```

```python
%%time
df = pak.fast_startswith( manystrings, 'U', 'found', search_me.V ) 
mask = df.found.notnull()
result1 = df[mask]   
#print( result0.shape[0],   result1.shape[0])
assert result0.shape[0] == result1.shape[0]

# result1: iterate through groups of strings with same length.
# also returns the matching fragment
# Wall time: 7.64 s (Notepad, size = 10000)
```

#### find_identical, find_longest

```python
# search self
df = pak.fast_startswith(search_me, 'V', 'found_long', search_me.V, find_identical=False) 
mask = df.found_long.notnull()
self1 = df[mask]   
#self1
```

```python
# search self
df = pak.fast_startswith(search_me, 'V', 'found_short', search_me.V, find_identical=False, find_longest=False) 
mask = df.found_short.notnull()
self2 = df[mask]   
#self2
```

```python
result = pd.merge(self1, self2, how='outer')
assert self1.shape[0] == result.shape[0]
assert self2.shape[0] == result.shape[0]
mask = result.found_short != result.found_long

result[mask]
```

# endswith

```python
# create data to search in
size = 1000
vornamen = pd.DataFrame(pak.random_series(1000,'name', name='V').drop_duplicates())
suche_nach = ['a','na','ina']
vornamen
```

```python
pak.analyse_cols(vornamen)
```

```python
# built-in endswith
mask = vornamen.V.str.endswith(tuple(suche_nach))
result0 = vornamen[mask]
result0.shape[0]
result0
```

```python
# pak.fast_endswith
df = pak.fast_endswith( vornamen, 'V', 'found_long', pd.Series(suche_nach) ) 
mask = df.found_long.notnull()
result1 = df[mask]   
assert result0.shape[0] == result1.shape[0]
result1
```

```python
pak.analyse_cols(result1)
```

```python
# fast_endswith mit find_longest=False
df = pak.fast_endswith( vornamen, 'V', 'found_short', pd.Series(suche_nach), find_longest= False ) 
mask = df.found_short.notnull()
result2 = df[mask]   
assert result0.shape[0] == result2.shape[0]
result2
```

```python
result = pd.merge(result1, result2, how='outer')
assert result1.shape[0] == result.shape[0]
assert result2.shape[0] == result.shape[0]
result
mask = result.found_short != result.found_long

result[mask]
```

```python
# Testdaten
anz = 100000
a = pak.random_series( anz, 'name',                                   p_nan=0,   p_dup=0, name='vorname')
b = pak.random_series( anz, 'int',    min=0,  max=2,                  p_nan=0,   p_dup=0, name='von')
c = pak.random_series( anz, 'int',    min=0,  max=5,                  p_nan=0,   p_dup=0, name='plus')
d = pak.random_series( anz, 'int',    min=-3, max=-1,                 p_nan=0,   p_dup=0, name='bis2')
df = pak.dataframe( (a, b, c, d) )
df['bis'] = df.von + df.plus
mask = (df['bis'] > df.vorname.str.len())   |   (df['bis'] == 0)
df.loc[mask,'bis'] = df[mask].bis2
df = pak.drop_cols(df,['plus','bis2'])
df
```

```python
pak.slice_string(df, 'vorname', 'von', 'bis', 'vorname_cut')
```
