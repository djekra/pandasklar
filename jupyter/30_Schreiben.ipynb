## Elementweise - apply auf einzelne Series

```python
# Die Funktion bekommt jeden einzelnen Wert der Series und kann ihn ändern
def ändere(skalar):
    return '#__' + skalar + '__#'
```

```python
# Wert für Wert anwenden
df = pak.leute()
mask = df['geburtsstadt'] == 'Bremen'
df.loc[mask, 'vorname'] = df.loc[mask, 'vorname'].apply(ändere)
df
```

## Zeilenweise - apply auf ganzen DataFrame
Vorsicht! Die Funktion wird ev. mehr als einmal ausgeführt!

```python
def ändere(zeile):
    zeile.alter = zeile.alter *2
    zeile.vorname = str.upper(zeile.vorname)
    return zeile    
```

```python
# Zeilenweise anwenden
df = pak.leute()
mask = df['geburtsstadt'] == 'Bremen'
df.loc[mask] = df.loc[mask].apply(ändere, axis=1)
df
```

```python
# mit selbstdefinierter Funktion
def generiere_zusammenfassung( zeile ):
    zusammenfassung = ''
    for spalte in zeile.index:
        zusammenfassung += spalte + '_' + str(zeile[spalte]) + ' '
    zeile['zusammenfassung'] = zusammenfassung.strip()
    return zeile
    
df = df.apply(generiere_zusammenfassung, axis=1)
df
```

## Zeilenweise ein neues Dataframe erzeugen

```python
# Testdaten
df = pak.leute(10)
erste_zeile = df.iloc[0]
df
```

```python
erste_zeile
```

```python
# Liefert ein DataFrame
def generiere(zeile):
    result = pd.DataFrame()
    for elt in zeile.history:
        list_list = [[zeile.vorname, zeile.alter, elt]]
        print('list_list:', list_list)
        neue_zeile = pd.DataFrame(list_list, columns=list('ABC')) # einzeiliges DataFrame    
        #result = result.append(neue_zeile, ignore_index=True)
        result = pd.concat( [result, neue_zeile], ignore_index=True)
    return result    
```

```python
# ausprobieren an der ersten Zeile
generiere(erste_zeile)
```

```python
# apply erzeugt eine Series von DataFrames
# tolist macht daraus eine Liste
r = df.apply(generiere, axis=1).tolist()
```

```python
# und concat macht aus der Liste von DataFrames ein Gesamt-DataFrame
pd.concat(r)
```
