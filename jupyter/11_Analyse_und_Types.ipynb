```python
# Zufallsdaten
anz = 10000
a = pak.random_series( anz, 'int',   min=-500, max=100 )
b = pak.random_series( anz, 'int',   min=-127, max=127, p_dup=0 ) # keine Dups erlaubt
c = a + b
a = a % 10 * 10
v = pak.random_series( anz, 'name',                  p_nan=0)
w = v.str[:1]
s = pak.random_series( anz, 'string',                p_nan=0)
t = pak.random_series( anz, 'string',                p_nan=0.1)
m = pak.random_series( anz, 'int',   min=0, max=127, p_nan=0.1 )
n = pak.random_series( anz, 'float',                 p_nan=0.2 ) #* 70000
o = pak.random_series( anz, 'choice', choice=['Bremen','Bremerhaven'], p_nan=0.3,   p_dup=0     )
p = pak.random_series( anz, 'list',                                    p_nan=0.1,   p_dup=0.5   )
z = pak.random_series( anz, 'mix',                                     p_nan=0.1,   p_dup=0     )

df = pak.dataframe( [a, b, c, v, w, s, t, m, n, o, p, z] )
df.columns = ['int_grob','int_fein','float_summe','Vorname','Anfangsbuchstabe','string_nonan','string_nan','int_nan','float_nan','Stadt','Liste','Mix']
df.float_summe = df.float_summe.astype('float')
```

## copy_datatype
Kopiert die dtypes von df_from auf df_to für alle Spaltennamen die übereinstimmen

```python
df1 = df.copy()                  # Falsche Datentypen
df1 = pak.drop_cols(df1,'int_fein')  # Ein Feld fehlt
df2 = pak.change_datatype(df)        # Richtige Datentypen
df2 = pak.drop_cols(df2,'Stadt')     # Ein Feld fehlt
```

```python
result = pak.copy_datatype(df1, df2)

# Ergebnis: Alle Spalten außer 'Stadt' wurden behandelt
pak.analyse_cols(result)
```

```python
# nochmal die vollständige Kopiervorlage
df2 = pak.change_datatype(df, verbose=False)         # Richtige Datentypen

# einzelnes Feld korrigieren
df1.Stadt = pak.copy_datatype(df1.Stadt, df2.Stadt)  
```

```python
# Ergebnis: Nur die Spalte 'Stadt' wurde behandelt
pak.analyse_cols(df1)
```

## pak.col_names
* selektiert Spaltennamen auf Basis der Analysen. Sinnvoll um eine Methode auf bestimmte Spalten eines DataFrame anzuwenden.

```python
?pak.col_names
```

```python
pak.col_names(df, only='int', query='nnan == 0 and ndups < 800', as_list=False )
```

```python
# Beispiel: fillna auf alle Sting-Spalten
df2 = df.copy()
spalten = pak.col_names(df2, only='str', query='nnan > 0')

print('Spalten:',spalten)
print('any_nan davor:', pak.any_nan(df2[spalten]))

# anwenden
df2[spalten] = df2[spalten].fillna('')
print('any_nan danach:', pak.any_nan(df2[spalten]))
```

```python
# Beispiel: astype auf alle int64-Spalten
# Zunächst: geeigneten Datentyp heraussuchen
pak.type_info('int8').info()
```

```python
# Beispiel: astype auf alle int64-Spalten
df2 = df.copy()
spalten = pak.col_names(df2, only='int', query='vmin > 0 and vmax < 127 and nnan == 0') # nur int_grob
spalten
```

```python
df2[spalten] = df2[spalten].astype( pak.type_info('int8').class_object  )
```

```python
spalten = pak.col_names(df2, only='int')
pak.analyse_cols(df2[spalten])
```

## Vergleich zweier Dataframes

```python
?pak.get_different_rows 
```

```python
# Zwei Spalten ausschließen
cols = pak.col_names(df, query='is_hashable')
df1 = df[cols].copy()
df2 = df[cols].copy()
```

```python
assert pak.get_different_rows(df1, df2).shape[0] == 0

df2.int_grob = df2.int_grob.apply( pak.decorate, p=0.003, special=42) # NaN wird übrigens auch erkannt
pak.get_different_rows(df1, df2)
```
