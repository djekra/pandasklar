# Work with NaN
* `nnan`: Count NaNs in Series or DataFrames.
* `any_nan`: Are there NaNs? Returns True or False.
* `nan_rows`: Returns the rows of a DataFrame that are NaN in the specified column.

```python
df = pak.people()
```

## nnan()


```python
?pak.nnan
```

```python
pak.nnan(df)
```

## any_nan()

```python
?pak.any_nan
```

```python
# any_nan
r1 = pak.any_nan(df)     
r2 = pak.any_nan(df.first_name)
r3 = pak.any_nan(df, without=['postal_code','birthplace'])

print(r1, r2, r3)
```
