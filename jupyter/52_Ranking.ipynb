```python
import numpy      as np
import pandas     as pd 
import bpyth      as bpy
import pandasklar as pak 

import time
from pandas._testing import assert_frame_equal
```

```python
# Testdaten erzeugen
np.random.seed(10)
anz = 1000 #0 #00
df = pak.zufallsdaten(anz)
df['AA'] = df.A.apply(lambda x: int(x/10)*10 )
df
```

## Standardmethode

```python
?pak.rank
```

```python
# Ranking-Parameter
#group_spalten = ['AA']     # Sehr wenige sehr große Gruppen
#group_spalten = ['A']      # Wenige große Gruppen
group_spalten = ['B']      # Viele kleine Gruppen
score_spalte  = 'D'
```

```python
# rank_sort_drop:
# schnell wenn wenige große Gruppen
# extrem schnell wenn viele kleine Gruppen
# Ergebnis ist unsortiert

t1 = time.time()

ranked3 = pak.rank(  df, group_spalten, score_spalte)

print('N =',anz)
print('Zeit =', round(time.time() - t1, 1 ), 'sek')
ranked3
```

## Alternative Methoden

```python
# rank_sort:
# ok wenn wenige große Gruppen
# ok wenn viele kleine Gruppen
# Ergebnis ist sortiert
t1 = time.time()

ranked1 = pak.rank_sort(      df, group_spalten, score_spalte) 

print('N =',anz)
print('Zeit =', round(time.time() - t1, 1 ), 'sek')
ranked1
```

```python
# rank_idxminmax:
# sehr schnell wenn wenige große Gruppen
# extrem langsam wenn viele kleine Gruppen
# Ergebnis ist sortiert
t1 = time.time()

ranked2 = pak.rank_idxminmax(  df, group_spalten, score_spalte)

print('N =',anz)
print('Zeit =', round(time.time() - t1, 1 ), 'sek')
ranked2
```

```python
# Identität der Ergebnisse prüfen
ranked3 = ranked3.sort_values(group_spalten)
assert_frame_equal(ranked1, ranked2)
assert_frame_equal(ranked2, ranked3)
```

# Rank markieren

```python
ranked_index = pak.rank(  df, ['A'], 'D').index
```

```python
result = df.copy()
result['best'] = False
result.loc[ranked_index, 'best'] = True
```

```python
result.sort_values(group_spalten + [score_spalte])
```

# Komplettes Ranking
statt nur das beste Ergebnis zu finden

```python
ranked4 = pak.rank(  df, group_spalten, score_spalte, richtung='max', target_spalte='ranking', method='first')
```

```python
ranked4.sort_values(group_spalten + [score_spalte], ascending=[True,False])
```

```python
pak.analyse_cols(ranked4)
```

```python
# Identität feststellen
mask = (ranked4.ranking==1)
spalten = ranked2.columns.values
ranked4 = ranked4[mask][spalten].sort_values(group_spalten + [score_spalte])
assert_frame_equal(ranked2, ranked4)
```

## Komplettes Ranking mit Konflikten

```python
group_spalten = ['AA']     # Sehr wenige sehr große Gruppen
score_spalte  = 'D'
```

```python
ranked5 = pak.rank(  df, group_spalten, score_spalte, richtung='max', target_spalte='ranking', method='average')
ranked5 = ranked5.sort_values(group_spalten + [score_spalte], ascending=[True,False])
ranked5
```
