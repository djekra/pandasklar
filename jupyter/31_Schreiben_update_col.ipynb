## Beispieldaten

```python
def beispieldaten(anz=10, nan_alter=True):
    
    # Beispieldaten
    np.random.seed(11)
    df = pak.leute(anz)
    df = pak.drop_cols(df,['merkmale','history'])
    
    # translatedaten
    # mit zwei NaNs
    tr = df.tail(int(anz/2))
    tr = pak.drop_cols(tr, ['geburtsstadt'])
    tr.passwort = '##' + tr.passwort + '##'
    tr.plz      = -tr.plz
    if nan_alter:
        tr.iloc[1,1] = np.NaN
    tr.iloc[3,4] = None
    tr = pak.reset_index(tr)
    return  df, tr

df,tr = beispieldaten()

df = pak.change_datatype(df, verbose=False)
df.alter = pak.change_datatype(df.alter, 'float32')
df = pak.rename_col(df, 'alter','aLtEr')
df
```

## pak.update_col
* Die Datensätze 5, 6, 7, 8, 9 sind in beiden DataFrames vorhanden. Aber:
* NaN oder None verhindern die Zuordnung, daher wird Datensatz 6 (Sarah) nicht geschrieben
* NaN oder None werden nicht übertragen,  daher wird Datensatz 8 (Pia)   nicht geschrieben
* Es bleiben also 5, 7, 9 (Sigrid, Arthur, Elena)

```python
df1 = df.copy()
df1 = pak.update_col( df1, tr, left_on=['vorname','aLtEr'], right_on=['vorname','alter'], col='passwort', keep='_alt')
pak.get_different_rows(df, df1)
```

```python
# tr auf Int wechseln, alter verändern
tr.alter = pak.change_datatype(tr.alter, 'Int16')
tr.alter += 100
tr.alter = tr.alter.fillna(12)
tr
```

```python
# dtype ist erhalten geblieben
# nicht jedes alter wurde überschrieben
df1 = pak.update_col(df1, tr, on='vorname', col='alter', col_rename='aLtEr', cond='min')
df1
```

## Mit mask

```python
df2 = df1.copy()
altersschwelle = int(df2.aLtEr.mean()+1)
print('altersschwelle',altersschwelle)
mask = (df2.aLtEr < altersschwelle)
df2[mask] = pak.update_col( df2[mask], tr, left_on=['vorname','aLtEr'], right_on=['vorname','alter'], col='plz')
```
